```python
import pandas as pd
from pandas_datareader import data as pdr
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sbn
import numpy as np

mglu_df = pd.read_csv('/content/Magazine Luiza MGLU3.csv')
ptr_df = pd.read_csv('/content/Petrobras PETR4.csv')

print('# Ações MGLU3')
print(mglu_df.head(10))

print('\n# Ações PTR4')
print(ptr_df.tail(10))
ptr_df.shape
```

```python
# Verificar dados ausentes ou duplicados em MGLU3
#mglu_df.count()
print("Qtde de valores ausentes - MGLU3:\n")
mglu_df.isnull().sum()
```

```python
# Verificar dados ausentes ou duplicados em PTR4
#ptr_df.count()
print("Qtde de valores ausentes - PTR4:\n")
ptr_df.isnull().sum()
```

```python
mglu_resumo = mglu_df[['DATA','FECHAMENTO', 'VARIAÇÃO', 'VOLUME']].copy()
ptr_resumo = ptr_df[['DATA','FECHAMENTO', 'VARIAÇÃO', 'VOLUME']].copy()

#Conveter coluna DATA para datetime
mglu_resumo['DATA'] = pd.to_datetime(ptr_resumo['DATA'],format="%d/%m/%Y")
ptr_resumo['DATA'] = pd.to_datetime(ptr_resumo['DATA'],format="%d/%m/%Y")

## MGLU3
#Converter colunas Fechamento e Variação em float
mglu_resumo['FECHAMENTO'] = mglu_resumo['FECHAMENTO'].str.replace(',','.').astype(float)
mglu_resumo['VARIAÇÃO'] = mglu_resumo['VARIAÇÃO'].str.replace(',','.').astype(float)
print('\nInformações em MGLU3: \n')
mglu_resumo.info()

## PTR4
#Converter colunas Fechamento e Variação em float
ptr_resumo['FECHAMENTO'] = ptr_resumo['FECHAMENTO'].str.replace(',','.').astype(float)
ptr_resumo['VARIAÇÃO'] = ptr_resumo['VARIAÇÃO'].str.replace(',','.').astype(float)
print('\nInformações em PTR4: \n')
ptr_resumo.info()
```

```python
#Renomear a coluna Volume para VOLUME (Em M e B)
ptr_resumo = ptr_resumo.rename(columns={'VOLUME':'VOLUME (Em M e B)'})
mglu_resumo = mglu_resumo.rename(columns={'VOLUME':'VOLUME (Em M e B)'})

ptr_resumo.head()
```

```python
#Remover referência B e M na coluna Volume - PTR4
ptr_resumo['VOLUME (Em M e B)'] = ptr_resumo['VOLUME (Em M e B)'].str.strip(" B")
ptr_resumo['VOLUME (Em M e B)'] = ptr_resumo['VOLUME (Em M e B)'].str.strip(" M")

#Remover referência B e M na coluna Volume - MGLU3
mglu_resumo['VOLUME (Em M e B)'] = mglu_resumo['VOLUME (Em M e B)'].str.strip(" B")
mglu_resumo['VOLUME (Em M e B)'] = mglu_resumo['VOLUME (Em M e B)'].str.strip(" M")

#Converter a coluna VOLUME (Em M e B) para float
ptr_resumo['VOLUME (Em M e B)'] = ptr_resumo['VOLUME (Em M e B)'].str.replace(',','.').astype(float)
mglu_resumo['VOLUME (Em M e B)'] = mglu_resumo['VOLUME (Em M e B)'].str.replace(',','.').astype(float)

mglu_resumo.info()
```

```python
#1. Qual foi a data com o menor e com o maior preço de ações nas duas empresas?
# MGLU3
preco_min_mglu = mglu_resumo['FECHAMENTO'].min()
data_min_mglu = mglu_resumo.loc[mglu_resumo['FECHAMENTO'].idxmin(),'DATA']
print(f'O menor fechamento identificado para a MGLU3 no período foi de: R$ {preco_min_mglu:.2f} no dia {data_min_mglu}')

preco_max_mglu = mglu_resumo['FECHAMENTO'].max()
data_max_mglu = mglu_resumo.loc[mglu_resumo['FECHAMENTO'].idxmax(),'DATA']
print(f'O maior fechamento identificado para a MGLU3 no período foi de: R$ {preco_max_mglu:.2f} no dia {data_max_mglu}\n')

# PTR4
preco_min_ptr = ptr_resumo['FECHAMENTO'].min()
data_min_ptr = ptr_resumo.loc[ptr_resumo['FECHAMENTO'].idxmin(),'DATA']
print(f'O menor fechamento identificado para a PTR4 no período foi de: R$ {preco_min_ptr:.2f} no dia {data_min_ptr}')

preco_max_ptr = ptr_resumo['FECHAMENTO'].max()
data_max_ptr = ptr_resumo.loc[ptr_resumo['FECHAMENTO'].idxmax(),'DATA']
print(f'O maior fechamento identificado para a PTR4 no período foi de: R$ {preco_max_ptr:.2f} no dia {data_max_ptr}\n')
```
